# Estimating $\pi$

Imagine that we inscribe a circle in a square of side $2a$ and pick a point uniformly at random from within the square. The probability that the selected point also belongs to the circle is the ratio of the areas of the two shapes: $\frac{\pi a^2}{(2a)^2} = \pi/4$.

```python
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
plt.figure(figsize=(12,12))
plt.plot([-1, 1], [1, 1], 'k', 
         [-1, 1], [-1, -1], 'k', 
         [-1, -1], [-1, 1], 'k', 
         [1, 1], [-1, 1], 'k')

x = np.linspace(-1, 1, 100)
y = (1 - x * x) ** 0.5

plt.plot(x, y, 'b', x, -y, 'b')

n = 10000
samples = zip(np.random.uniform(low=-1, high=1, size=n), np.random.uniform(low=-1, high=1, size=n))
count = 0

for x,y in samples:
    if x**2 + y**2 <= 1:
        plt.scatter(x, y, c='g')
        count += 1
    else:
        plt.scatter(x, y, c='r')

plt.show()
pi = 4 * count / n
print("Estimated value of pie: {}".format(pi))
```
