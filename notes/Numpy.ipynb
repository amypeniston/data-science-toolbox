# Numpy

* Powerful and efficient array operations
* Save time by not using loops to apply mathematical operations
* Avoids type-checking at runtime = faster (note: all elements must be of the same type)
* Saves memory as elements are not pointers to disparate data but rather contiguous in storage

```python
photo = io.imread("../assets/city.jpg")
plt.imshow(photo)
```

```python
plt.imshow(photo[200:400,0:200])
```

```python
plt.imshow(photo[::-1]) # start:stop:step
```

```python
plt.imshow(photo[::-1, ::-1]) # flip vertically & horizontally
```

#### Masks

```python
c = np.array([3, 1, 5, 9])
c > 3
```

```python
c[c > 3]
```

```python
masked = np.where(photo > 100, 255, 0) # condition, if true, if false
plt.imshow(masked)
```
