# Python - Visualization

The most widely used data visualization library is ```matplotlib```. It was originally created by John Hunter, a neurobiologist, to visualize ECoGs. Much like MATLAB, ```matplotlib``` makes it very easy for non-programmers to quickly and easily generate graphics.

The architecture of ```matplotlib``` is composed of three layers:

**The back-end layer** is comprised of three built-in low level classes: FigureCanvas, Renderer and Event.

**The artist layer** contains one main object: Artist. All visual elements are an Artist instance (either primitive or composite type). The axes are the most important instance of type composite because that is where most of the API plotting methods are defined. 

**The scripting layer** was developed for non-programmers to perform quick exploratory data analysis. Users interact with this layer via the ```matplotlib.pyplot``` interface.

````python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches # needed for waffle charts
from PIL import Image # to convert images into arrays
import seaborn as sns
````

#### Box Plots

A box plot is a statistical representation of the distribution of the data:

* **Minimum** - the smallest value in the dataset
* **First Quartile** - 25% value; midpoint between minimum and median
* **Median** - middle value of the distribution
* **Third Quartiles** - 75% value; midpoint between median and maximum
* **Maximum** - the largest value in the dataset

The range between the first and third quartiles is called the **Inter-Quartile Range (IQR)**, where IQR = Q3 - Q1.

Also included on a box plot are **outliers**, which are:

* Greater than Q3 by at least 1.5 time the IQR
* Less than Q1 by by at least 1.5 time the IQR

To see the exact values used on a box plot, use the method ```df.describe()```.

#### Scatter Plots

````python
# Extract total immigration per year
df_tot = pd.DataFrame(df[years].sum(axis=0))
df_tot.index = map(int, df_tot.index)
df_tot.reset_index(inplace = True)
df_tot.columns = ['year', 'total']

# Fit the data
x = df_tot['year']
y = df_tot['total']
fit = np.polyfit(x, y, deg=1)

# Plot the data
df_tot.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='darkblue')
plt.title('Total Immigration to Canad, 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

# Plot line of best fit
plt.plot(x, fit[0] * x + fit[1], color='red')
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 150000))

plt.show()
````

#### Bubble Plots

A bubble plot is a variation on a scatter plot that displays three dimensions of data, (x, y, z), where the size of the data points are determined by z (also known as the "weight"). We can create these weights using a normalization technique such as feature scaling.

````python
# Prepare the data
df_t = df[years].transpose()
df_t.index = map(int, df_t.index)
df_t.index.name = 'Year'
df_t.reset_index(inplace=True)


norm_china = (df_t["China"] - df_t["China"].min())/(df_t["China"].max() - df_t["China"].min())
norm_india = (df_t["India"] - df_t["India"].min())/(df_t["India"].max() - df_t["India"].min())

ax0 = df_t.plot(kind="scatter",
                    x = "Year",
                    y = "China",
                    s = norm_china * 2000 + 10,
                    color = "blue",
                    alpha = 0.5,
                    figsize=(15,5)
                   )

ax1 = df_t.plot(kind="scatter",
                    x = "Year",
                    y = "India",
                    s = norm_india * 2000 + 10,
                    color = "green",
                    alpha = 0.5,
                    ax = ax0
                   )

plt.title("Total # of Immigrants from China & India, 1980 - 2013")
plt.ylabel("# of Immigrants")
plt.legend(["China", "India"], loc="upper left", fontsize="x-large")
plt.show()
````

#### Subplots

Subplots enable us to view multiple visualizations within the same figure. When working with subplots, we typically use the Artist layer, rather than the scripting layer, as it gives us more control over placement.

First, create a figure representing the overall canvas. Then, divide this figure into multiple subplots, assigning each to its own axes variable.

````python
fig = plt.figure() # create figure

ax0 = fig.add_subplot(1, 2, 1) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(1, 2, 2) # add subplot 2 (1 row, 2 columns, second plot)

# Subplot 1: Box plot
df_CI.plot(kind='box', color='blue', vert=False, figsize=(20, 6), ax=ax0) # assign to subplot 1
ax0.set_title('Box Plots of Immigrants from China and India, 1980 - 2013')
ax0.set_xlabel('Number of Immigrants')
ax0.set_ylabel('Countries')

# Subplot 2: Line plot
df_CI.plot(kind='line', figsize=(20, 6), ax=ax1) # assign to subplot 2
ax1.set_title ('Line Plots of Immigrants from China and India, 1980 - 2013')
ax1.set_ylabel('Number of Immigrants')
ax1.set_xlabel('Years')

plt.show()
````

#### Waffle Charts

A waffle chart is a unique visualization tool that is often used to display progress towards a goal or relative proportions of different categories.

````python
def create_waffle_chart(categories, values, height, width, colormap, value_sign=''):
    # Determine relative proportions
    overall_total = sum(values)
    category_proportions = [float(value / overall_total) for value in values]

    # Set up waffle
    total_num_tiles = width * height

    # Computer # of tiles per category
    tiles_per_category = [round(proportion * total_num_tiles) for proportion in category_proportions]

    # Initial waffle as empty matrix
    waffle_chart = np.zeros((height, width))

    # Define indices to loop through waffle chart
    category_index = 0
    tile_index = 0

    # Populate waffle chart
    for col in range(width):
        for row in range(height):
            tile_index += 1

            # Check if num tiles puts us into the next category
            if tile_index > sum(tiles_per_category[0:category_index]):
                category_index += 1

            waffle_chart[row, col] = category_index

    # Create a visual
    fig = plt.figure()
    plt.matshow(waffle_chart, cmap=colormap)
    plt.colorbar()

    # Set/style ticks
    ax = plt.gca()
    ax.set_xticks(np.arange(-0.5, (width), 1), minor=True)
    ax.set_yticks(np.arange(-0.5, (height), 1), minor=True)
    ax.grid(which="minor", color='w', linestyle='-', linewidth=2)
    plt.xticks([])
    plt.yticks([])

    # Compute cumsum to match color schemes of chart & legend
    values_cumsum = np.cumsum(values)
    total_values = values_cumsum[len(values_cumsum) - 1]

    # Create legend
    legend_handles = []
    for i, category in enumerate(categories):
        if value_sign == '%':
            label_str = category + "(" + str(values[i]) + values_sign + ")"
        else:
            label_str = category + "(" + value_sign + str(values[i]) + ")"
        color_val = colormap(float(values_cumsum[i])/overall_total)
        legend_handles.append(mpatches.Patch(color=color_val, label=label_str))

    # Add legend
    plt.legend(handles=legend_handles,
              loc="lower center",
              ncol=len(categories),
              bbox_to_anchor=(0., -0.2, 0.95, 1))

    plt.show()
````

````python
width = 40
height = 10

df_dns = df.loc[["Denmark", "Norway", "Sweden"]]

categories = df_dns.index.values
values = df_dns["Total"]

colormap = plt.cm.viridis

create_waffle_chart(categories, values, height, width, colormap)
````

#### Word Clouds

Word clouds are a simplistic way to visualize the content of textual data. Words that appear more often in the source will appear bigger and bolder in the corresponding word cloud. A Python package called ```word_cloud```, which was developed by Andreas Mueller, makes it easy to generate these visualizations.

````python
# install wordcloud
!pip install wordcloud

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS
````

````python
# Load text
war_of_the_worlds = open("../assets/war-of-the-worlds.txt", 'r').read()
````

````python
# Create a unique list of stopwords that will be excluded
stopwords = set(STOPWORDS)
stopwords.add("said") # Add said to the stopwords

# Instantiate a word cloud object
word_cloud = WordCloud(
    background_color="white",
    max_words=2000,
    stopwords=stopwords
)
word_cloud.generate(war_of_the_worlds)

# Display the word cloud
fig = plt.figure(figsize=(15,5))
plt.imshow(word_cloud, interpolation="bilinear")
plt.axis("off")
plt.show()
````

````python
spaceship_mask = np.array(Image.open("assets/spaceship.jpg"))
plt.imshow(spaceship_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
````

````python
# Instantiate a word cloud object
word_cloud = WordCloud(
    background_color="white",
    max_words=2000,
    stopwords=stopwords,
    mask=spaceship_mask
)
word_cloud.generate(war_of_the_worlds)

# Display the word cloud
fig = plt.figure(figsize=(15,5))
plt.imshow(word_cloud, interpolation="bilinear")
plt.axis("off")
plt.show()
````

````python
# https://data.sfgov.org/Public-Safety/Police-Department-Incident-Reports-2018-to-Present/wg3w-h783
# Load SF Police Report dataset
police_df = pd.read_csv("../assets/Police_Department_Incident_Reports__2018_to_Present.csv")
````

````python
incidents_df = police_df[['Incident Datetime', 'Incident Date', 'Incident Time', 'Incident Year',
       'Incident Day of Week', 'Report Datetime', 'Row ID','Incident Number','Filed Online', 'Incident Code',
       'Incident Category', 'Incident Subcategory','Resolution', 'Police District','Analysis Neighborhood',
       'Latitude', 'Longitude','point']]
# Remove incidents that do not have a lat/lng
incidents_df = incidents_df.dropna(subset=["Latitude","Longitude"])

# Only take the first 100 incidents
incidents_df = incidents_df.iloc[0:100, :]
````

````python
incidents = folium.map.FeatureGroup()
san_fran = folium.Map(location=[37.77, -122.42], zoom_start=12)

# Add circles to each incident location
for lat, lng in zip(incidents_df.Latitude, incidents_df.Longitude):
    incidents.add_child(
        folium.vector_layers.CircleMarker(
            [lat, lng],
            radius=3,
            color="red",
            fill=True,
            fill_color="yellow",
            fill_opacity=0.4
        )
    )

# Add markers with popups
latitudes = list(incidents_df.Latitude)
longitudes = list(incidents_df.Longitude)
labels = list(incidents_df["Incident Category"])

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(san_fran)

san_fran.add_child(incidents)
````

#### Chloropleth Maps

In Chloropleth maps, areas are shaded in proportion to the measurement of the statistical variable that is being displayed. They are a great way to visualize how a measurement changes over different geographic regions; typically, the higher the value, the darker the region.

The basis of any chloropleth map is a GeoJSON file that defines the areas/boundaries of each region that we are interested in.

````python
world_geo = r"../assets/world_countries.json"
df.reset_index(inplace=True)
````

````python
df.head()
````

````python
world_map = folium.Map(location=[0,0], zoom_start=2, tiles="Mapbox Bright")

folium.Choropleth(
    geo_data=world_geo,
    data=df,
    columns=["Country", "Total"],
    key_on="feature.properties.name", # Per GeoJSON file
    fill_color="YlGn",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Immigration to Canada"
).add_to(world_map)

world_map
````
